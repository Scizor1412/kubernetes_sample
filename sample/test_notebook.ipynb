```python
# Given an integer array nums and an integer target, return the indices of the two numbers such that they add up to target.
 
def get_substract(nums, target):
    result_set = set()
    for index, value in enumerate(nums):
        substract = target - value
        if substract in nums:
            result_set.add(tuple(sorted((index, nums.index(substract)))))
    return result_set
        
nums = [2, 7, 5, 4]
target = 9
# Output: [0, 1]

get_substract(nums, target)
```

```python
# Write a function called print_list_indexes() that takes 2 parameters:
# String
# an arbitrarily complex nested list
# The function should print the value of each list element on a separate line. The value of each line should be preceded by the string and numbers indicating the depth and index of the element in the list. Assume that the list contains only strings and other nested lists.
# Input:
# my_list = ['h', 'a', ['e','b'], 'c']
# arb_str = 'Foo'
# Output:
# Foo:0 h
# Foo:1 a
# Foo.2:0 e
# Foo.2:1 b
# Foo:3 c
def iterate(arb_str, nested_list, layer=0, index_str=''):
    for index, value in enumerate(nested_list):
        if isinstance(value, list):
            layer += 1
            index_str += str(index) + ':'
            iterate(arb_str, value, layer, index_str)
        else:
            if layer == 0:
                print(arb_str + ':' + str(index) + ' ' + value)
            else:
                print(arb_str + '.' + index_str + str(index) + ' ' + value)

my_list = ['h', 'a', ['e','b', ['oaoisd', 'laksjl', 'lkx']], 'c']
arb_str = 'Foo'
iterate(arb_str, my_list)
```

```python
# Given a string s, write a function to determine if the string is a palindrome, considering only alphanumeric characters and ignoring cases.
 
# Example:
# Input: s = "A man, a plan, a canal: Panama"
# Output: True
 
# Input: s = "race a car"
# Output: False

a = 'race a car'

def is_palindrome(input_string):
    reformatted_list = [i.lower() for i in input_string if i.isalnum()]
    reverse_formatted_list = reformatted_list.copy()
    reverse_formatted_list.reverse()
    if reformatted_list == reverse_formatted_list:
        print(True)
    else:
        print(False)

is_palindrome(a)
```

```python
import math

# validate input data
def validate_input(input_number):
    try:
        result = float(input_number)
        return result
    except:
        return 0

# dictionary for unit names
unit_to_dict = {
    1: 'thousand',
    2: 'million',
    3: 'billion',
    4: 'trillion',
    5: 'quadrillion'
}

# add a function to convert 3 digit number to word
def convert_3_digit_number_to_word(shortened_number):
    hundreds = int(shortened_number)//100
    tens = int(shortened_number) % 100

    number_to_word = {
        1: 'one',
        2: 'two',
        3: 'three',
        4: 'four',
        5: 'five',
        6: 'six',
        7: 'seven',
        8: 'eight',
        9: 'nine'
    }

    tens_to_word = {
        2: 'twenty',
        3: 'thirty',
        4: 'forty',
        5: 'fifty',
        6: 'sixty',
        7: 'seventy',
        8: 'eighty',
        9: 'ninety',
        10: 'ten',
        11: 'eleven',
        12: 'twelve',
        13: 'thirteen',
        14: 'fourteen',
        15: 'fifteen',
        16: 'sixteen',
        17: 'seventeen',
        18: 'eighteen',
        19: 'nineteen'
    }

    words = []
    if hundreds > 0:
        words.append(number_to_word[hundreds])
        if hundreds > 1:
            words.append('hundreds')  
        else: 
            words.append('hundred')
    if tens > 0:
        if tens < 20 and tens > 9:
            words.append(tens_to_word[tens])
        elif tens < 10:
            words.append(number_to_word[tens])
        else:
            tens_place = tens // 10
            units_place = tens % 10
            words.append(tens_to_word[tens_place])
            if units_place > 0:
                words.append(number_to_word[units_place])
    return words

# convert the input number to parts similar to human language
def convert_number_to_words(input_number):
    output = []
    input_length = len(str(input_number))
    for i in range(math.ceil(input_length/3)):
        unit = unit_to_dict.get(math.floor(input_length/3) - i, '')
        if i == 0 and input_length%3 == 0:
            continue
        elif i == 0:
            words = convert_3_digit_number_to_word(input_number[:input_length%3])          
        else:
            words = convert_3_digit_number_to_word(input_number[input_length%3 + i*3 - 3: input_length%3 + i*3])
            
        if words:
            output.extend(words)
            output.append(unit)
    return output


# main function
if __name__ == '__main__':
    # input
    input_number = input('Enter a number: ')
    output = []
    # validate input
    if validate_input(input_number) == 0:
        print('Invalid number format. Please input correct currency format.')
    else:
        # convert whole part to words
        whole_part = input_number.split('.')[0]
        whole_length = len(whole_part)
        print('Input number:', input_number)
        print('Whole part:', whole_part)
        output = convert_number_to_words(whole_part)
        output.append('dollars' if int(whole_part) > 1 else 'dollar')
        if '.' in input_number:
            decimal_part = input_number.split('.')[1]
            if len(decimal_part) > 2:
                decimal_part = decimal_part[:2]
            elif len(decimal_part) == 1:
                decimal_part = decimal_part + '0'
            decimal_part = int(decimal_part)
            print('Decimal part:', decimal_part)
            decimal_part_to_word = convert_3_digit_number_to_word(decimal_part)
            output.extend(decimal_part_to_word)
            output.append('cent')
    print(output)
    print(' '.join(output))
```

```python
# Question:
# Write a Python function "is_anagram_set()" that compares two strings and returns TRUE if they are anagrams of each other. An anagram is a word formed by rearranging the letters of another word. For example "cat" and "act" are anagrams because they both contain the same letters "A", "C", and "T". But "cat" and "tact" are NOT anagrams because there is an extra "T" in "tact". All of the letters from one word must be used to form the other word.

def is_anagram_set(str1, str2):
    str1_list = sorted([i.lower() for i in str1 if i.isalpha()])
    str2_list = sorted([i.lower() for i in str2 if i.isalpha()])

    return True if str1_list == str2_list else False


is_anagram_set('cat', 'alskdj')
```

```python
# The game is played on a standard 10x10 board
# Player 1 starts on position 0x0 (top left corner). Player 2 starts on position 9x9 (bottom right corner)
# At each turn, a player may move up, down, left, or right on the board.
# These steps are held in an array and take the form ‘u’, ‘d’, ‘l’ and ‘r’ respectively
# If a player crosses a previous path of another player, including themselves, they are eliminated
# It a player lands on the sake space as another player on the same turn, both players are eliminated and the match is declared a draw
# If a player moves off the board, into the vast cyber nothingness, they are eliminated.
# If there is only one player left at the end of a turn, that player wins no matter if they have sore moves or not.
# If the match has ended and there is more than one player still active, the match is declared a draw,
# If both players are eliminated on the same turn, the match is declared a draw.
map_size = 10
a = {'x': 0, 'y': 0, 'icon': 'first'}
b = {'x': map_size - 1, 'y': map_size - 1, 'icon': 'second'}
moves = ['u', 'd', 'l', 'r']

his_moves = {'f': [(0,0)], 's': [(map_size - 1, map_size - 1)]}

def move_player(player):
    false_count = 3
    while false_count > 0:
        move = input(f"Player {player['icon']}'s turn (use u, d, l and r for the player to move accordingly up, down, left, or right on the board): ")
        if move.lower() not in moves or len(move):
            print('You must use u, d, l and r to move the player!')
            false_count -= 1
        else:
            if move == 'u':
                player['y'] -= 1
            elif move == 'd':
                player['y'] += 1
            elif move == 'l':
                player['x'] -= 1
            elif move == 'r':
                player['x'] += 1

            his_moves[player['icon']].append((player['x'], player['y']))
            return True
    if false_count == 0:
        print('You have used wrong input 3 times. The game will end now!')
        return False

def map_drawer(first_player, second_player, map_size):
    for i in range(map_size):
        for j in range(map_size):
            if i == first_player['y'] and j == first_player['x']:
                print('x', end='')
            elif i == second_player['y'] and j == second_player['x']:
                print('y', end='')
            else:
                print('*', end='')
        print('')
    print('')

def check_player_exist(player):
    if player['x'] < 0 or player['x'] > (map_size -1) or player['y'] < 0 or player['y'] > (map_size -1):
        return False
    else:
        return True

def check_player_cross_path(player):
    other_his_moves = []
    for i in his_moves:
        if i == player['icon']:
            continue
        else:
            other_his_moves.extend(his_moves[i])

    if (player['x'], player['y']) in other_his_moves:
        return False
    else:
        return True

while True:
    map_drawer(a, b, map_size)

    if move_player(a) == False or check_player_cross_path(a) == True:
        print(f"Player {a['icon']} has been eliminated. Player {b['icon']} is the winner!")
        break

    map_drawer(a, b, map_size)

    if move_player(b) == False or check_player_cross_path(b) == True:
        print(f"Player {b['icon']} has been eliminated. Player {a['icon']} is the winner!")
        break

    map_drawer(a, b, map_size)

    if check_player_exist(a) and check_player_exist(b):
        if a == b:
            print(f'2 players are standing in the same place!. The game is draw!')
        else:
            continue
    elif check_player_exist(a) and not check_player_exist(b):
        print(f"Player {b['icon']} has left the map, player {a['icon']} is the winner!")
        break
    elif check_player_exist(b) and not check_player_exist(a):
        print(f"Player {a['icon']} has left the map, player {b['icon']} is the winner")
        break
    else:
        print("Both player no longer in the map, the game is draw!")
        break
    


```
